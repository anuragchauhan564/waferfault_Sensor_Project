#### # Get Columns to Drop:

```python
def get_cols_with_zero_dev(df : pd.DataFrame):
    """
    Returns a list of columns names who are having zero standard deviation.
    """
    cols_to_drop = []
    num_cols = [col for col in df.columns if df[col].dtype != 'O'] # numerical cols only
    for col in num_cols:
        if df[col].std() == 0:
            cols_to_drop.append(col)
    return cols_to_drop

def get_redundant_cols(df: pd.DataFrame, missing_thresh = .7):
    """Return a list of columns having missing values more than certain thresh
    """
    cols_missing_ratio = df.isna().sum().div(df.shape[0])
    cols_to_drop = list(cols_missing_ratio[cols_missing_ratio > missing_thresh].index)
    return cols_to_drop
```

```python
# Columns with missing vals more than .7
cols_to_drop_1 = get_redundant_cols(wafers, missing_thresh=.7)
cols_to_drop_1
```

```python
## Columns with 0 Standard Deviation 

cols_to_drop_2 = get_cols_with_zero_dev(df=wafers)
cols_to_drop_2
```

```python
##  Columns to Drop 
cols_to_drop = cols_to_drop_1 + cols_to_drop_2
```

###### => Features that are not gonna contribute to ML algorithm in anyway
